```python
#--------------------- Importation du dictionnaire et random ----------------------#

dico=open("liste_francais.txt","r").read().split() # liste des mots

import random


#--------------------- Choix du mot par l'ordinateur ----------------------#

mot=random.choice(dico) # mot donné par l'odinateur pris dans la liste de mot

#--------------------- Définition des variables ----------------------#

nbreLettreTrv=(len(mot)-1)

US=" _ " #underscore

motATRV=[mot[0]+US*nbreLettreTrv] # mot du départ 

erreur=0 # nombre d'erreur que l'utilisateur a faite

LettresUB=[" Voici les lettres qui sont correctes et qui ont été entrées : "] #  lettres entrée par le joueur qui sont bonnes

taille=['_']*(len(mot)) # le nombre de lettre a trouver qui est remplacée par des underscores pour l'instant

PenduDessinTEMP=[] # dessin du pendu temporaire

LettresUB=[" Voici les lettres qui sont correctes et qui ont été entrées : "] #  lettres entrée par le joueur qui sont bonnes

taille=['_']*(len(mot)) # le nombre de lettre a trouver qui est remplacée par des underscores pour l'instant

PenduDessinTEMP=[] # dessin du pendu temporaire

LettresUF=[" Voici les lettres qui sont fausses et qui ont été entrées : "] #  lettre entrée par le joueur qui sont fausses


erreurmax=11 

PenduDessin=["---------","|| \n || \n || \n || \n || \n---------","|| / \n ||/ \n || \n || \n || \n---------","----------- \n || / \n ||/ \n || \n || \n || \n---------","----------|- \n || /      |  \n ||/ \n || \n || \n || \n---------","----------|- \n || /      |  \n ||/       o  \n || \n || \n || \n---------","----------|- \n || /      |  \n ||/       o  \n ||        | \n || \n || \n---------","----------|- \n || /      |  \n ||/       o  \n ||       /| \n || \n || \n---------","----------|- \n || /      |  \n ||/       o  \n ||       /|\ \n || \n || \n---------","----------|- \n || /      |  \n ||/       o  \n ||       /|\ \n ||       / \n || \n---------","----------|- \n || /      |  \n ||/       o  \n ||       /|\ \n ||       / \\n || \n ---------"]


del(taille[0])     # Clara: la fontcion del je l'ai trouvé sur : https://fr.moonbooks.org/Articles/Comment-supprimer-une-élément-dune-liste-avec-python/

#---------------------- Affichage ----------------------#

print("Les règles du jeu de ce pendu sont très simples:\n\n1. L'odinateur va donner un mot au hasard, et vous allez devoir deviner toutes les lettres ce mot !\n2. Vous allez devoir entrer une lettre, si elle fait partie du mot alors elle remplacera le _ correspondant, dans le cas contraire le dessin d’un pendu se met à apparaître.\n3. Attention:\n             -Il ne faudra pas re entrez une lettre qui a déjà été saisis et qui est fausse, parce que cela vous ajoutera juste une étape de plus au dessin du pendu donc moins de chance de trouver le mot.\n             - La première lettre du mot n'est pas considérer comme une lettre déjà saisis il faudra alors la saisir si vous le voulez, mais c'est un risque à prendre car si elle ne fait pas partie du mot, alors cette action sera comptée comme fausse et donc cela va vous enlever le droit à une erreur\n             - L'orthographe, les lettres à trouver dans cette liste de mots ne sont que des lettres en miniscule veuillez ne saisir que des lettres miniscules; les traits d'union compte comme des lettre à trouver si il y en a; les lettres avec des accents etc... sont qualifiées comme une lettre (par exemples é, et è sont considérées comme deux lettres distinctes). Exemples:é,è,ê,ë,à,ï,ô,ç, etc...")# Regles du jeu

#print("\n",mot)

print("\nLe nombre de lettres à trouver dans le mot est de",nbreLettreTrv)
        
print(motATRV)

#--------------------- Jeu ---------------------#


while not erreur==11: 
    
    Lettre=input("\n \n Quelle lettre pensez vous qu'il y a dans le mot donné par l'ordinateur ? : ")
    
    if Lettre in mot:
        if Lettre in LettresUB:
            print("\n Cette lettre est correcte mais elle a deja été entrée ")
      
        else:
            LettresUB.append(Lettre) #ajt la lettre dite par l'utilisateeur au tableau LettresUB si elle n'en fait pas partit
            print("\n",LettresUB)
    
        
        
        # Toute cette fonction permet de remplacer les underscore par les lettres trouvées avec details d'affichage 
        for i in range (len(taille)+1): 

            if Lettre == mot[i]:   
                taille[i-1]=Lettre
                print("\n",mot[0],taille)
                
                if (len(PenduDessinTEMP))!=0:
                    print(PenduDessinTEMP)  
                
        if "_" not in taille:
              print(f"Vous avez gagné!! avec seulement {erreur} erreurs ! Le mot était donc bien : {mot}")
              break # permet de mettre fin au jeu quand le joueur a trouvé le mot
    elif not Lettre in mot:
        
        if not Lettre in LettresUB:
            LettresUF.append(Lettre) #ajt la lettre dite par l'utilisateeur au tableau LettresUB si elle n'en fait pas partit
            print("\n",LettresUF)
            
        erreurmax-=1
        print(f"\n Cette lettre ne fait pas parti du mot à trouver, il ne vous reste que le droit a : {erreurmax} erreurs")
        erreur+=1
        print("\n",PenduDessin[erreur-1]) # affiche le dessin du pendu correspondant au nombre d'erreurs 
        PenduDessinTEMP=PenduDessin[erreur-1]
        
        
print("\n Vous avez perdu, le pendu est fini, vous avez atteint le maximum d'erreurs possible. Le mot était ",mot,". Dommage re essayez ! ")
```
