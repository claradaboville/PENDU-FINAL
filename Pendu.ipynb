```python
dico=open("liste_francais.txt","r").read().split() # liste des mots

#print(dico[:100])

import random




#--------------------- Définition ----------------------#

mot=random.choice(dico) # mot donné par l'odinateur pris dans la liste de mot

nbreL=len(mot)
nbreLettreTrv=int(len(mot)-1) # nombre de lettre à trouver dans le mot, -1 car la première lettre est déjà donné

underscore=" _ "
UnderS=[underscore]*nbreLettreTrv


motATRV=[mot[0],UnderS] # mot du départ à trouver 

occurrence=0

erreur=0
erreurmax=11
taille=['_']*(len(mot))


del(taille[0])     # la fontcion del je l'ai trouvé sur : https://fr.moonbooks.org/Articles/Comment-supprimer-une-élément-dune-liste-avec-python/
#--------------------- Définition ----------------------#

#---------------------- Affichage ----------------------#

print("Les règles du jeu de ce pendu sont très simples:\n\n1. L'odinateur va donner un mot au hasard, et vous allez devoir deviner toutes les lettres ce mot !\n2. Vous allez devoir entrer une lettre, si elle fait partie du mot alors elle remplacera le _ correspondant, dans le cas contraire le dessin d’un pendu se met à apparaître.\n3. Attention:\n             -Il ne faudra pas re entrez une lettre qui a déjà été saisis et qui est fausse, parce que cela vous ajoutera juste une étape de plus au dessin du pendu donc moins de chance de trouver le mot.\n             - La première lettre du mot n'est pas considérer comme une lettre déjà saisis il faudra alors la saisir si vous le voulez, mais c'est un risque à prendre car si elle ne fait pas partie du mot, alors cette action sera comptée comme fausse et donc cela va vous enlever le droit à une erreur\n             - L'orthographe, les mots dans cette liste ne sont que des mots en miniscule veuillez ne saisir que des lettres miniscules; les lettres avec des accents etc... sont qualifiées comme une lettre (par exemples é, et è sont considérées comme deux lettres distinctes). Exemples:é,è,ê,ë,à,ï,ô,ç, etc...")# Regles du jeu

print("\n",mot)

print("\nLe nombre de lettres à trouver dans le mot est de",nbreLettreTrv)

print(motATRV)
#---------------------- Affichage ----------------------#
#--------------------- Boucle ---------------------#


while not erreur==11:
    l=input("Quelle lettre pensez vous qu'il y a dans le mot donné par l'ordinateur ? : ")
    if l in mot:
        
        for i in range (len(taille)):
            if l == mot[i]:
                taille[i-1]=l
                print(mot[0],taille)
              
           
    else:
        erreurmax-=1
        print("Cette lettre ne fait pas parti du mot à trouver, il ne vous reste que le droit a ",erreurmax)
        erreur+=1
    
    
```
